```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/amazon_products_with_images (1).csv")

# Required columns
df = df[['product_name', 'category', 'brand', 'price']].copy()

# Remove missing values
df = df.dropna()

# Convert price to numeric
df['price'] = (
    df['price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

df['price'] = pd.to_numeric(df['price'], errors='coerce')

df = df.dropna()

# -----------------------------
# User Input
# -----------------------------

search = input("Enter Product Name : ").lower()

# Search product
product = df[df['product_name'].str.lower().str.contains(search)]

if product.empty:
    print("Product not found!")
    exit()

selected = product.iloc[0]

selected_price = selected['price']

print("\nSelected Product")
print("-----------------------------")
print("Product :", selected['product_name'])
print("Brand   :", selected['brand'])
print("Price   : ₹", selected_price)

# Price range (±10%)
lower = selected_price * 0.90
upper = selected_price * 1.10

recommend = df[
    (df['price'] >= lower) &
    (df['price'] <= upper) &
    (df['product_name'] != selected['product_name'])
]

# Sort by closest price
recommend['difference'] = abs(recommend['price'] - selected_price)
recommend = recommend.sort_values('difference')

print("\nTop 5 Similar Price Products")
print("-----------------------------")

for _, row in recommend.head(5).iterrows():
    print("Product :", row['product_name'])
    print("Brand   :", row['brand'])
    print("Category:", row['category'])
    print("Price   : ₹", row['price'])
    print("-----------------------------------")
```
