```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import NearestNeighbors

df = pd.read_csv("/content/amazon_products_with_images (1).csv")

df = df[['product_name','category','rating','price']].dropna()

encoder = LabelEncoder()
df['category']=encoder.fit_transform(df['category'])

X=df[['category','rating','price']]

model=NearestNeighbors(n_neighbors=6)

model.fit(X)

distances,indices=model.kneighbors([X.iloc[0]])

print("Selected Product\n")
print(df.iloc[0]['product_name'])

print("\nRecommended Products\n")

for i in indices[0][1:]:
    print(df.iloc[i]['product_name'])
```
